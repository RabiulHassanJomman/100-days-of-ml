```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_science_job.csv')
df.head()
```

```python
df.isnull().sum()/len(df) * 100
```

```python
df.isnull().mean()*100
```

```python
cols = [var for var in df.columns if df[var].isnull().mean() < 0.05 and df[var].isnull().mean() > 0]
cols
```

```python
df[cols].sample(5)
```

```python
df['education_level'].value_counts()
```

```python
len(df[cols].dropna())/len(df)
```

```python
new_df = df[cols].dropna()
df.shape, new_df.shape
```

```python
new_df.hist(bins=50, density=True, figsize=(12, 12))
plt.show()
```

```python
new_df.hist(bins=50, density=False, figsize=(12, 12))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['training_hours'].hist(bins=20, ax = ax, density = True, color = 'red')
new_df['training_hours'].hist(bins=20, ax = ax, density = True, color = 'green', alpha=0.9)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['training_hours'].plot.density(color='red')
new_df['training_hours'].plot.density(color='green', alpha=.5)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

df['city_development_index'].hist(bins=20, ax = ax, density = True, color = 'red')
new_df['city_development_index'].hist(bins=20, ax = ax, density = True, color = 'green', alpha=0.9)
```

```python
df['city_development_index'].plot.density(color='red')
new_df['city_development_index'].plot.density(color='green')
```
