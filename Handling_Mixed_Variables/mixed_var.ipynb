```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('passengers travelling with')
```

```python
df['number_numeric'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
df['number_cat'] = np.where(df['number_numeric'].isnull(), df['number'], np.nan)
```

```python
df.head()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(/d+)')
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df.head()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
df['ticket_numeric'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_numeric'] = pd.to_numeric(df['ticket_numeric'], errors='coerce', downcast='integer')
```

```python
df.head()
```

```python
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])
```
